```python
import numpy as np
import gym
import random
```

```python
env = gym.make("Taxi-v2")
env.render()
```

```python
action_size = env.action_space.n
print(action_size)

s_s = env.observation_space.n
print(s_s)
```

```python
qtable = np.zeros((s_s, action_size))
print(qtable)
```

```python
total_ep = 50000
total_test_ep = 100
max_steps = 99

lr = 0.7
g = 0.618
eps = 1.0
max_eps = 1.0
min_eps = 0.1
decay_rate = 0.01
```

```python
for ep in range(total_ep):
    s = env.reset()
    step = 0
    done = False
    
    for step in range(max_steps):
        exp_exp_tradeoff = random.uniform(0,1)
        
        if exp_exp_tradeoff > eps:
            action = np.argmax(qtable[s, :])
        
        else:
            action = env.action_space.sample()
            
        new_state, reward, done, info = env.step(action)
        qtable[s, action] = qtable[s, action] + lr * (reward + g * np.max(qtable[new_state, :]) - qtable[s, action])
        
        s = new_state
        
        if done == True:
            break
    
    ep += 1
    
    eps = min_eps + (max_eps - min_eps) * np.exp(-decay_rate * ep)
    
        
```

```python
env.reset()
rewards = []

for ep in range(total_test_ep):
    s = env.reset()
    step = 0
    done = False
    total_rewards = 0
    print("******************************************")
    print("EPISODE", ep)
    
    for step in range(max_steps):
        env.render()
        action = np.argmax(qtable[s,:])
        new_state, reward, done, info = env.step(action)
        total_rewards += reward
        
        if done:
            rewards.append(total_rewards)
            print("Score", total_rewards)
            break
        s = new_state
env.close
print("Score over time" + str(sum(rewards)/total_test_ep))
```
